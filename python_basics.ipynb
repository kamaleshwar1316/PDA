#FOR LOOP

```python
for i in range(1, 6):
    print(i)
```

#WHILE LOOP

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

#BREAK

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

#CONTINUE STATEMENT

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

#PASS STATEMENT

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
```

#MATCH CASE

```python
day = 2

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid day")
```
